```python
import pandas as pd
import re
import urllib.request
import naas
```

```python
#data = pd.read_csv('CallbackDetail_ayeT_expDate_Sat Nov 12 2022 21_08_18 GMT+0700 (Western Indonesia Time).csv', sep=';')
data = pd.read_html('https://us-central1-dopami.cloudfunctions.net/exportAyetData')
data
```

```python
data = data[0]
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data
```

```python
data = data.drop_duplicates()
```

```python
#data['timestamp'] = data['timestamp'].str.replace(r'\s\w+\w+\+\w+\s\(\w+\s\w+\)',' ')
data['timestamp'] = data['timestamp'].str.replace(r'\s\w+\w+\+\w+\s\(\w+\s\w+\s\w+\)',' ')
data
```

```python
data_new = data
data_new['timestamp'] = pd.to_datetime(data_new['timestamp'])
data_time_sorted = data_new.sort_values('timestamp', axis=0)
data_time_sorted_con = data_time_sorted[::-1]
data_time_sorted
```

```python
data_time_sorted_con
```

```python
date_filtered = data_time_sorted['timestamp'].diff().tolist()
date_filtered
```

```python
date_filtered_con = data_time_sorted_con['timestamp'].diff().tolist()
date_filtered_con
```

```python
x_time = [abs(x.total_seconds()/60) for x in date_filtered]
x_time
```

```python
x_time2 = [abs(x.total_seconds()/60) for x in date_filtered_con]
x_time2
```

```python
data_time_sorted['Timedelta'] = x_time
data2 = data_time_sorted
data2
```

```python
data_time_sorted_con['Timedelta'] = x_time2
data3 = data_time_sorted_con
data3
```

```python
data2['Timedelta'] = data2['Timedelta'].astype('float')
#data2['Timedelta'].info()
```

```python
data_clean_delta = data2.loc[data2['Timedelta'] < 10.0]
data_clean_delta
```

```python
data3['Timedelta'] = data3['Timedelta'].astype('float')
#data3['Timedelta'].info()
```

```python
data_clean_delta2 = data3.loc[data3['Timedelta'] < 10.0]
data_clean_delta2
```

```python
frames = [data_clean_delta, data_clean_delta2]
clean_data_final = pd.concat(frames)
clean_data_final = clean_data_final.drop(columns=['Timedelta'])
clean_data_final
```

```python
clean_data_final = clean_data_final.drop_duplicates()
clean_data_final
```

```python
pattern_uid = r"\w+"
```

```python
data_uid = clean_data_final['uid']
```

```python
fr_uid = []
for i, uid in enumerate(data_uid):
    for j, uid2 in enumerate(data_uid[i+1:]):
        try:
            check_uid = re.search(pattern_uid, uid)
            check_uid2 = re.search(pattern_uid, uid2)
        except KeyError:
            pass
        else:
            if check_uid.group() == check_uid2.group():
                fr_uid.append(uid)
                fr_uid.append(uid2)
```

```python
new2_data = clean_data_final.loc[clean_data_final['uid'].isin(fr_uid)]
new2_data = new2_data.sort_index()
new2_data
```

```python
unique_uid = new2_data['uid'].unique()
unique_uid
```

```python
idx = []
for uid_new in unique_uid:
    idx.append(new2_data.loc[new2_data.UID==uid_new,'uid'].index[-1])

new2_data = new2_data.drop(idx)
new2_data
```

```python
new2_data['amount'] = new2_data['amount'].astype('int')
```

```python
hasil = abs(new2_data.groupby('uid')['amount'].sum())
hasil
```

```python
hasil_index = hasil.loc[hasil > 10000].index.values.tolist()
hasil_index #FRAUD
```

```python
list_link = []
for uid_fraud_final in hasil_index:
    list_link.append('https://us-central1-dopami.cloudfunctions.net/pointsCallback?uid='+uid_fraud_final+'&point=-50')
list_link
```

```python
for pengurangan in list_link:
    urllib.request.urlopen(pengurangan)
```
