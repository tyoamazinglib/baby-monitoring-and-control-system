```python
import pandas as pd
import numpy as np
import re
import urllib.request
from datetime import date
```

```python
data = pd.read_html('https://us-central1-dopami.cloudfunctions.net/exportAdgemData')
data
```

```python
data = data[0]
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data
```

```python
data['conversion_datetime'] = data['conversion_datetime'].str.replace(r'\d+\:\d+\:\d+\s\w+\+\d+\s\(\w+\s\w+\s\w+\)','')
data['conversion_datetime'] = data['conversion_datetime'].str.replace(r'^\w+','')
data
```

```python
today = date.today()
day = today.strftime("%b-%d-%Y").replace('-', ' ')
day
# day = 'Sep 17 2022'
```

```python
data_time_1 = data['conversion_datetime']
data_time_1
```

```python
pat = r'\w+\s\w+\s\w+'
```

```python
data_now = []
data_past = []
for i, time in enumerate(data_time_1):
    try:
        check_time = re.search(pat, time)
        check_now = re.search(pat, day)
    except KeyError:
        pass
    else:
        if check_time.group() == check_now.group():
            data_now.append(time)     
        else:
            data_past.append(time)   
```

```python
dataframe_now = data.loc[data['conversion_datetime'].isin(data_now)]
dataframe_now
```

```python
dataframe_past = data.loc[data['conversion_datetime'].isin(data_past)]
dataframe_past
```
